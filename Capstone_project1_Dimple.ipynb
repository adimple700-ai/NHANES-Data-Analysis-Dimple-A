# NHANES Data Analysis Project - Dimple A

This notebook analyzes NHANES adult body measurement data using Python. It includes data loading, cleaning, visualization, BMI calculation, waist-to-height ratio comparison, and standardized output.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Step 1: Connecting to NHANES data...")
male_url = "https://raw.githubusercontent.com/gagolews/teaching-data/master/marek/nhanes_adult_male_bmx_2020.csv"
female_url = "https://raw.githubusercontent.com/gagolews/teaching-data/master/marek/nhanes_adult_female_bmx_2020.csv"

# Load data and clean it
male_df = pd.read_csv(male_url).dropna()
female_df = pd.read_csv(female_url).dropna()
male = male_df.values
female = female_df.values

print("Step 2: Generating Weight Comparison...")
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(female[:, 0], bins=20, color='pink', edgecolor='black')
plt.title('Female Weights')
plt.xlabel('Weight')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(male[:, 0], bins=20, color='skyblue', edgecolor='black')
plt.title('Male Weights')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

print("Step 3: Calculating BMI...")
f_bmi = female[:, 0] / ((female[:, 1] / 100) ** 2)
female = np.column_stack((female, f_bmi))
zfemale = (female - np.mean(female, axis=0)) / np.std(female, axis=0)

print("Step 4: Body Ratios (Waist-to-Height)...")
f_ratio = female[:, 6] / female[:, 1]
m_ratio = male[:, 6] / male[:, 1]
plt.figure(figsize=(7, 5))
plt.boxplot([f_ratio, m_ratio], labels=['Female Ratio', 'Male Ratio'])
plt.title('Waist-to-Height Comparison')
plt.ylabel('Waist-to-Height Ratio')
plt.show()

print("Step 5: Final Standardized Data (Top 5 BMI):")
sorted_idx = np.argsort(female[:, 7])
print(pd.DataFrame(zfemale[sorted_idx[-5:]], columns=list(male_df.columns)+['BMI']))

print("\n--- DONE! You can now share your link! ---")
```
